```python
import os
root = 'data/summary_bart'

for word_number in [30, 50, 100]:
    holder = []
    for summary_bart in list(sorted(filter(lambda x: f'_{word_number}' in x, os.listdir(root)), key=lambda x: int(x.split('_')[0]))):
        with open(root + f'/{summary_bart}') as file:
            holder.append(file.readline())

    data[f'bart_summary_{word_number}'] = holder
data
```
